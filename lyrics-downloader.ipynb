```python
# scrape page
url = 'https://www.azlyrics.com/lyrics/nas/lifesabitch.html'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

# extract artist, song title
details = soup.find_all('b')
song = details[1].text[1:-1]
artist = details[0].text[:-7]

# extract album and year
album_info = soup.find_all('div', class_='songinalbum_title')
album_info = list(album_info)[0].text
album_regex = r'"(.*)"'
year_regex = r'\((\d*)\)'
album = re.findall(album_regex, album_info)[0]
year = re.findall(year_regex, album_info)[0]

# extracrt lyrics
html = list(soup.children)[2]
body = list(html.children)[3]
lyrics = list(list(list(body.children)[9])[3])[5].find_all('div')[5].text

# add entry to df
lyrics_df.loc[1] = [artist, album, year, song, lyrics]
```
